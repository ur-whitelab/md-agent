```python
protein_name = 'thermophilic serine proteinase'
pdb = '1DBI'
```

```python
model = 'gpt-4o-2024-05-13'
top_k_tools = 'all'

import datetime
from mdagent import MDAgent

agent_params = {
        "agent_type": "Structured",
        "model": model,
        "tools_model": model,
        "use_memory": False,
        "top_k_tools": top_k_tools,
}
mdagent = MDAgent(**agent_params)

now = datetime.datetime.now()
date = now.strftime("%Y-%m-%d")
time = now.strftime("%H:%M:%S")
print("date and time:",date,time)
print("LLM: ",mdagent.llm.model_name,"\nTemperature: ",mdagent.llm.temperature)
```

```python
prompt = (
    "Some proteins are not stable under normal conditions. "
    "From literature, "
    f"determine the approriate temperature for {protein_name} with PDB ID {pdb}. "
    "Finally, choose the most correct (closest) answer below: \n"
    "\t A. 2.0  \t B. 5.0  \t C. 6.0  \t D. 7.0  \t E. 8.5 \n"
)
print(prompt)
```

```python
mdagent.run(prompt)
```

```python
assert any(tool.name != 'GetOptimalConditions' for tool in mdagent.tools), "Tool named 'GetOptimalConditions' found in the list!"
assert any(tool.name == 'LiteratureSearch' for tool in mdagent.tools), "No tool named 'LiteratureSearch' found in the list."
```
