```python
model = 'gpt-4o-2024-05-13'
top_k_tools = 'all'

import datetime
from mdagent import MDAgent

agent_params = {
        "agent_type": "Structured",
        "model": model,
        "tools_model": model,
        "use_memory": False,
        "top_k_tools": top_k_tools,
}
mdagent = MDAgent(**agent_params)

# ran on July 22 at roughly 5 or 6pm

# now = datetime.datetime.now()
# date = now.strftime("%Y-%m-%d")
# time = now.strftime("%H:%M:%S")
# print("date and time:",date,time)
# print("LLM: ",mdagent.llm.model_name,"\nTemperature: ",mdagent.llm.temperature)
```

```python
prompt = (
    "Some proteins are not stable under normal conditions. "
    "From literature, find out about the stability of the protein and "
    f"determine the optimal pH for {protein_name} with PDB ID {pdb}. "

    "If no papers are found, try different known names for the protein. "
    "For example, keywords for procollagen C-endopeptidase enhancer 1 "
    "would be 'procollagen C-protease enhancer', 'type 1 procollagen'"

    "Finally, choose the most correct (closest) answer below: \n"
    "\t A. 2.0  \t B. 5.0  \t C. 6.0  \t D. 7.0  \t E. 8.5 \n"
)
print(prompt)
```
