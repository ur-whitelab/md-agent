```python
protein_name = 'alkaline phosphatase PafA'
pdb = '5TJ3'
```

```python
model = 'gpt-4o-2024-05-13'
top_k_tools = 'all'

from mdagent import MDAgent

agent_params = {
        "agent_type": "Structured",
        "model": model,
        "tools_model": model,
        "use_memory": False,
        "top_k_tools": top_k_tools,
}
mdagent = MDAgent(**agent_params)
```

```python
prompt = (
    "Some proteins are not stable under normal conditions. "
    "From literature, "
    f"determine the approriate temperature for {protein_name} with PDB ID {pdb}. "
    "Finally, choose the most correct (closest) answer below (all in degree C): \n"
    "\t A. 10  \t B. 25  \t C. 45  \t D. 60  \t E. 75 \n"
)
print(prompt)
```

```python
mdagent.run(prompt)
```
