```python
protein_name = 'alpha-amylase'
pdb = '3VM5'
```

```python
model = 'gpt-4o-2024-05-13'
top_k_tools = 'all'

import datetime
from mdagent import MDAgent

agent_params = {
        "agent_type": "Structured",
        "model": model,
        "tools_model": model,
        "use_memory": False,
        "top_k_tools": top_k_tools,
}
mdagent = MDAgent(**agent_params)

now = datetime.datetime.now()
date = now.strftime("%Y-%m-%d")
time = now.strftime("%H:%M:%S")
print("date and time:",date,time)
print("LLM: ",mdagent.llm.model_name,"\nTemperature: ",mdagent.llm.temperature)
```

```python
prompt = (
    "Some proteins are not stable under normal conditions. "
    "From the best of your knowledge alone, "
    f"determine the appropriate temperature for {protein_name} with PDB ID {pdb}. "
    "Choose the most correct (closest) answer below (all in degree C): \n"
    "\t A. 10  \t B. 25  \t C. 35  \t D. 50  \t E. 75 \n"
)
print(prompt)
```

```python
mdagent.run(prompt)
```

```python
print('Number of tools:',len(mdagent.tools))
for tool in mdagent.tools:
    print(tool.name)
```
