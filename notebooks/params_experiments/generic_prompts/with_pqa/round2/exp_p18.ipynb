```python
# run MDAgent to do parameter filling experiment

from mdagent import MDAgent, Evaluator

agent_params = {
        "agent_type": agent_type,
        "model": model,
        "use_memory": use_memory,
        "top_k_tools": top_k_tools,
        "ckpt_dir": ckpt_dir
}
mdagent = MDAgent(**agent_params)
```

```python
# NOTE: this prompt tests parameter filling and whether the simulation won't break
# there's another test prompt where additional params are timestep & number of steps, then plot RMSD over time. Not tested here. 

list_params_options =  ("Here's a list of parameters with options in parenthesis if applicable: \n preprocessing (standard cleaning, remove water, "
    "add hydrogens, add hydrogens and remove water. Default: add hydrogens and remove water), \n whether to solvate or not, "
    "\n force fields (default: Amber14), \n ensemble (NPT, NVT, NVE), \n integrator (LangevinMiddle, Verlet, Brownian), " 
    "\n temperature  in Kelvin, \n pressure, \n friction, "
    "\n non-bonded interaction method (NoCutoff, CutoffNonPeriodic, CutoffPeriodic, Ewald, PME, LJPME) and non-bonded cutoff, "
    "\n constraints (None, AllBonds, HBonds, HAngles) and constraint tolerance. \n\n")

prompt = (
    f"Use Literature Search tool, what input parameters should I use to simulate {material}? Example application may be "
    f"to study {application}. If you get 'I cannot answer' response from that tool, refine your "
    f"question by removing PDB ID or giving a list of some parameters and try again. "
    f"Don't give all parameters at once or don't ask for one by one. {list_params_options} "
    "After the search, make the list of parameters you chose and tell me which ones Literature Search helped you with. Summarize the "
    "explanation for these parameters under 50 words. "
    "Lastly, get the PDB file and simulate this molecular system with these parameters for 100 steps. "
)
print(prompt)
```
