```python
from mdagent import MDAgent
import nest_asyncio
nest_asyncio.apply()
```

```python
prompt6 = "Simulate 1A3N and 7VDE (two PDB IDs matching hemoglobin) with identical parameters. Find the appropriate parameters for simulating hemoglobin from the literature. Then, plot the radius of gyration throughout both simulations."
llm_var = "claude-3-5-sonnet-20240620"
tools = "all"
agent = MDAgent(agent_type="Structured", model=llm_var, top_k_tools=tools, use_memory=False)
print(llm_var)
```

```python
agent.run(prompt6)
```

```python
registry = agent.path_registry
paths_and_descriptions = registry.list_path_names_and_descriptions()
print("\n".join(paths_and_descriptions.split(",")))
```

```python
from IPython.display import Image
import re
import os
matches = re.findall(r"fig0_\d+", paths_and_descriptions)
```

```python
file_id1 = matches[0]
path1 = registry.get_mapped_path(file_id1)
assert os.path.exists(path1)
Image(filename=path1)
```

```python
file_id2 = matches[1]
path2 = registry.get_mapped_path(file_id2)
assert os.path.exists(path2)
Image(filename=path2)
```
