```python
import datetime
import os
from mdagent import MDAgent
```

```python
prompt2 = "Download the PDB file for protein 1LYZ."
llm_var = "accounts/fireworks/models/llama-v3p1-70b-instruct"
tools = "all"
agent = MDAgent(agent_type="Structured", model=llm_var, top_k_tools=tools)
now = datetime.datetime.now()
date = now.strftime("%Y-%m-%d")
print("date and time:",date)
time = now.strftime("%H:%M:%S")
print("time:",time)
print("LLM: ",agent.llm.model_name,"\nTemperature: ",agent.llm.temperature)
```

```python
agent.run(prompt2)
```

```python
now = datetime.datetime.now()
date = now.strftime("%Y-%m-%d")
print("date and time:",date)
time = now.strftime("%H:%M:%S")
print("time:",time)
```

```python
registry = agent.path_registry
print(os.path.basename(registry.ckpt_dir))
```

```python
paths_and_descriptions = registry.list_path_names_and_descriptions()
print("\n".join(paths_and_descriptions.split(",")))
```

```python
import re
import os
match = re.search(r"1LYZ_\d+", paths_and_descriptions)
file_id = match.group(0)
pdb_path = registry.get_mapped_path(file_id)
assert os.path.exists(pdb_path)
print(f'It is asserted that file path for {file_id} exists')
```
